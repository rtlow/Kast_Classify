```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd

#setting matplotlib parameters
matplotlib.rc('xtick', labelsize=14) 
matplotlib.rc('ytick', labelsize=14) 

#set matplotlib global font size
matplotlib.rcParams['font.size']=14
```

```python
df = pd.read_csv('/home/ryan/projects/data_prods/Kast_classify/Kastclassify_spectra_calculations_20200602-2.csv')
```

```python
df
```

```python
CaH1 = df['lepine2003_CaH1'].values
CaH2 = df['lepine2003_CaH2'].values
CaH3 = df['lepine2003_CaH3'].values
TiO5 = df['lepine2003_TiO5'].values



spt = df['subtype'].values
```

```python
CaH3
```

```python
CaH1_mod = []
for val in CaH1:
    try:
        res = float(val)
    except:
        res = np.nan
    CaH1_mod.append(res)
    
CaH1 = np.array(CaH1_mod)

CaH1
```

```python
TiO5
```

```python
fig, ax = plt.subplots(1,1, figsize=[7,7])
#ax.set_title('Reproduction of Fig. 2 in Lepine (2003)')
ax.scatter(TiO5, CaH3)
ax.set_xlim(0, 1.1)
ax.set_ylim(0,1.1)
ax.set_ylabel('CaH3')
ax.set_xlabel('TiO5')
fig.savefig('Lepine_TiO5_relation_2.png')
```

```python
fig, ax = plt.subplots(1,1, figsize=[7,7])
#ax.set_title('Reproduction of Fig. 2 in Lepine (2003)')
ax.scatter(TiO5, CaH2)
ax.set_xlim(0, 1.1)
ax.set_ylim(0,1.1)
ax.set_ylabel('CaH2')
fig.savefig('Lepine_TiO5_relation_1.png')
```

```python
fig, ax = plt.subplots(1,1, figsize=[7,7])
ax.set_title('Reproduction of Fig. 2 in Lepine (2003)')
ax.scatter(TiO5, CaH1)
ax.set_xlim(0, 1.1)
ax.set_ylim(0,1.1)
ax.set_ylabel('CaH1')
fig.savefig('Lepine_TiO5_relation_0.png')
```

```python
fig, ax = plt.subplots(3,1, figsize=[7,21])

ax[0].set_title('Reproduction of Fig. 2 in Lepine (2003)')
ax[0].scatter(TiO5, CaH1)
ax[0].set_xlim(0, 1.1)
ax[0].set_ylim(0,1.1)
ax[0].set_ylabel('CaH1')

ax[1].scatter(TiO5, CaH2)
ax[1].set_xlim(0, 1.1)
ax[1].set_ylim(0,1.1)
ax[1].set_ylabel('CaH2')

ax[2].scatter(TiO5, CaH3)
ax[2].set_xlim(0, 1.1)
ax[2].set_ylim(0,1.1)
ax[2].set_ylabel('CaH3')
ax[2].set_xlabel('TiO5')

fig.savefig('Lepine_TiO5_relation.png')

plt.show()
```
